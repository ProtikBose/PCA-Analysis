```python
# Library Function
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
import pandas as pd
from sklearn.preprocessing import StandardScaler
#print("Hello")
```

```python
# access data from csv
dataset = pd.read_csv('Data.csv')

#dataset.head()
```

```python
# data preprocessing 
saID=dataset["Sample ID"]
dataset=dataset.drop("Sample ID",axis=1)
dataset=dataset.drop("Source",axis=1)
# all columns 
col=dataset.columns
#dataset.head()
```

```python
X=np.array(dataset)
'''
scaler = StandardScaler()
scaler.fit(X)
X=scaler.transform(X) 

'''
```

```python
pca = PCA()
x_new = pca.fit_transform(X)
#print(x_new.shape)
#print(x_new)
#x_new
```

```python
def myplot(score,coeff,labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    
    # figure size
    fig=plt.figure(figsize=(20,9))
    
    # axis limitation
    plt.xlim(-1,1)
    plt.ylim(-1,1)
    
    # axis label
    plt.xlabel("F1".format(1))
    plt.ylabel("F2".format(2))
    
    # grid
    plt.grid()  
    
    
    
    # plotting data
    plt.scatter(xs * scalex*2,ys * scaley*3,c = "Red")
    
    # house number assigned
    for i, txt in enumerate(saID):
        plt.annotate(txt, (xs[i] * scalex*2,ys[i] * scaley*3))
    
    # arrow for columns
    for i in range(n):
        plt.arrow(0, 0, coeff[i,0]*2, coeff[i,1]*3,color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 2, coeff[i,1] * 1.5, col[i], color = 'g', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
        
    plt.show()
    # save png file in the local pc
    fig.savefig("pic.png",dpi=fig.dpi)

myplot(x_new[:,0:2],np.transpose(pca.components_[0:2, :]))
```
